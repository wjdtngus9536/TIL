## 2.0. The software process
2.1.1. Many different software processes but all involve
- Specification – defining what the system should **do**;
- Design and implementation – defining the organization of the system and implementing the system;
- Validation – checking that it does what the **customer wants**;
    - include verification
    - Done
- Evolution – changing the system in response to changing customer needs.
    - A/S

2.1.2. Plan-driven and agile processes

In agile processes, planning is incremental and it is
easier to change the process to **reflect changing
customer requirements.** 고객의 요구사항을 얼마나 잘 반영할 것인가  

In practice, most practical processes include elements of
both plan-driven and agile approaches.

### Incremental development : development and validation are interleaved


- The cost of accommodating changing customer requirements is reduced.

ncremental development problems
The process is not visible.
 Managers need regular deliverables to measure progress. If
systems are developed quickly, it is not cost-effective to produce
documents that reflect every version of the system.
 System structure tends to degrade as new increments
are added.
 Unless time and money is spent on refactoring to improve the
software, regular change tends to corrupt its structure.
Incorporating further software changes becomes increasingly
difficult and costly.

## 2.3. Coping with change_40분대 다시듣기
Change is inevitable in all large software projects.
- 사업 변화가 시스템 요구사항 변화를 만든다.
- 새 기술이 implementations의 개선을 가능하게 한다.
- 플랫폼 변화는 application 변화를 요구한다.
> Change leads to rework so the costs of change include
both rework (e.g. re-analysing requirements) as well as
the costs of implementing new functionality

비용과 재작업 줄이기
- 변화 예측, 
    - 발표할 때 구멍을 남겨두고 묻고싶게끔 만들어둔다.
- Change tolerance

Coping with changing requirements

- Software prototyping : 간단하게 보여주려고, 대부분 UI
    - The requirements engineering process to help with **requirements elicitation and validation;**
    - In design processes to explore options and develop a UI design;
    - In the testing process to run back-to-back tests.
    - Benefits:
        1. Improved system usability 사용성 개선
        2. A closer match to users’ real needs. 실 사용자의 니즈에 근접
        3. Improved design quality 디자인 퀄리티
        4. Improved maintainability 
        5. Reduced development effort 
- Incremental delivery
