## 3. 강의목표 및 평가방법
---

- 소프트웨어공학적 이론을 학습하여 전문적인 소프트웨어개발자로 성장하는 것을 목표로 한다.
- 팀원과의 소통과 맡은 임무에 책임을 다하여 프로젝트를 성공적으로 마치는 것을 목표로 한다.
- 개발과정을 managing하는 법
- **신뢰할 수 있는 프로그램이 더 중요**

- 중간시험 : 35%
- 기말시험 : 35%
- 기타(출석, 프로젝트 수행 등) : 30%


## 4. 소프트웨어란
---

- 프로그램과 프로글매의 개발, 운용, 수정 및 기능 확장을 위해 필요한 모든 정보
- 소프트웨어는 프로그램의 동적인 의미
    - 프로그램은 형식언어(formal language)로 표현된 `지적 구축물`
- 특징
    - 비가시성(invisiblity)
        - 개념적, 무형적, 구조가 코드 내에 내재
    - 복잡성(Complexiity)
        - 개발 과정이 복잡, 대상 업무 또는 시스템이 난해
    - 순응성(Conformity)
        - 요구나 환경 변화에 적절히 변형 가능
    - 복제 가능성(Duplicability)
    - 테스팅의 어려움(Intestability)
    - 변경 가능성(Changeability)
        - 미리 알아야한다.

## 7. 소프트웨어에 대한 오해 - 관리자
---
- 소프트웨어 개발에 관한 좋은 책들이 있고 책 안에 개발 표준과 단계가 제시되어 있어 우리에게 필요한 모든 것을 제공할 것이다.

- 개발자들에게 필요한 최신 기계나 CASE도구를 도입하였으니 좋은 제품을 빠른 시일 내에 만들 수 있을 것이다.

- 엔지니어들이 요구 분석을 하고 있으면 생산적이지 못한 일을 하고 있다고 생각한다.

- 공정이 지연될 경우 인력을 더 투입하면 해결된다.


## 8. 소프트웨어에 대한 오해 - 고객
---
- 목표에 대한 개략적인 기술만 해놓으면 소프트웨어를 만드는데 충분하다. 세부적인 것은 나중에 채워 넣으면 된다.

- 사용자의 요구 사항은 계속 변하며, 소프트웨어는 융통성이 있어 쉽게 변경을 수용할 수 있다.


## 9. 소프트웨어 위기
---

- 소프트웨어 개발의 문제
    - 예산 초과
    - 개발 일정 지연
    - 불충분한 성능
    - 신뢰하기 어려운 품질
    - 유지 보수의 어려움
    - 매우 비싼 유지 보수 비용
    
- 소프트웨어 위기
    - 대중화로 인해 소프트웨어의 수요가 급증하였으나, 소프트웨어의 생산성과 생산 기술은 그에 미치지 못하여 나타난 현상
    
## 10. 소프트웨어 공학

- 품질 좋은 소프트웨어를 최소의 비용으로 계획된 일정에 맞추어 개발하기 위하여 여러가지 공학적 원리와 방법을 체계적으로 적용하는 것

- Planning, Why build the system?
- Analysis, Who, What, When, Where, will the system be?
- Design, How will the system work?
- Implementation, System delivery
