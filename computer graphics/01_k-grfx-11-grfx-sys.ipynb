## 2. 래스터 시스템 Rester System

### 그래픽스 시스템
 - 이미지는 `프레임버퍼(화면에 나오는 부분을 그대로 저장하는 것들)`에 저장 
 
### 래스터 시스템(이미지 저장 방법)
 - 이미지 -> (색깔 있는) 사각형의 2차원 배열 = 2D array of colored rectangles
 - 픽셀 pixel = picture element = a single colored rectangle in the image

### 래스터 시스템 용어들
 - 픽셀 pixel = picture element
 - 래스터 raster = `픽셀의 2차원 배열` a 2D array of pixels
 - 프레임 버퍼 framebuffer(래스터 방식으로 저장된 것 중에 가장 중요한 것)  = 픽셀을 저장한 `메모리 영역`
     - 우리가 쓰는 그래픽 카드에는 반드시 프레임버퍼가 들어가 있음.
     
### 프레임버퍼의 기능
 - 프레임버퍼 -> 화면으로 **1:1 대응 관계**
 - 우리는 프레임 버퍼의 2차원 배열에 있는 내용을 바꿀 것.
 - 이게 하드웨어 적으로 구현됨
