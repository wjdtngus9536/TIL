평균, 표준편차, 분산 같은걸 구해서 대표값을 꼽는다거나 10개의 데이터 중 대표 데이터를 추출하여 
이것을 기반으로 알고리즘을 만들어 결과를 해석
결과가 좋지 않으면 특성 추출을 다시한다. (반복)
규칙 기반, 알고리즘이 만들어진 것
이미지 측면에서 특성을 추출한다는 것 : 필터를 씌우는 것 blur처리 하거나 특정한 이미지 필터를 만들어서 적용하여 나온 결과를 가지고
모델에 넣어서 머신러닝 기법 같은 것을 활용했었는데 모든 것을 다 담을 수 없고, 데이터의 특징을 제일 잘 아는 사람은 만든 사람일텐데
만든 사람이 그 데이터를 가지고 머신러닝이나 딥러닝 같은 것들을 활용하기에는 어려움이 많았다.

- 분류를 한다거나 수치를 계산해서 결과를 내뱉음
- 그 결과를 가지고 사람이 평가를 하면 된다
- 점점 사람이 할 수 없는 분야까지 확장하게 된 것

### 딥러닝의 기본 수식 y = w * x + b
그래프 표현


x : 입력 데이터
y : 결과값
w : 가중치
b : 편향

#### w(가중치)

w를 랜덤하게 시작한다. 0.1이 될 수도 있고 1이 될 수도 있고
Loss란? : y값과 모델 예측 값의 차이 e.g., 모델이 5라고 예측했는데 y가 4면 loss는 -1
모델 학습의 목표 : loss가 가장 작을 수 있게 하는 것

100개 중에 80개를 맞췄으면 80% : 절대적인 판단 기준인 100%가 있다
Loss는 수치만 봐서는 알 수 없음, 우리가 예측 해야하는 값의 범위가 어디까지인지 모르니까
모델의 결과값이 0 ~ 1사이로 나온다고 하면, 1과 0.9의 오차는 0.1
0 ~ 10사이에서 10과 예측값 9.9의 차이도 0.1 후자가 훨씬 예측을 잘한것

loss가 작아질수록 성능이 좋아지고 있다고 할 수 있음
0.2 >> 0.1 >> 0.09 라면 성능의 진척도가 더뎌졌다고 본다.

Loss가 가장 낮은 지점이 가장 좋은 지점

우리가 이 지점을 찾아서 학습을 멈춰야 한다.
모델 개발자들이 이렇게 코드를 짜줘야 한다.

w를 수정할 것인데 왼쪽으로 갈지 오른쪽으로 갈지 잘 모르니까

w를 왼쪽으로 조금 간 후에 loss를 다시 계산해본다.
loss가 낮아진 것을 보고 왼쪽으로 수정하니까 낮아진다는 것을 판단
얼마만큼 낮아진건지도 판단할 수 있을 것이다.
계속 이 방향으로 가게 될 것.

lr(learning rate) =  학습율, 얼만큼 이동하게 할 것인지
loss를 구하는 법도 천차만별 >> 하나의 수치를 보고 모델을 수정해야 하는 것

## 평가 측면

여전히 2차원 그래프
- x축 : Epoch
- y축 : Loss

많이 반복할 수록 나아질 수 있다.
하지만 특이점이 있을 것(몇 달씩 학습시키는 것은 비효율적)
평행에 가기 시작하면 그만하는 것이 적절

모델이 학습했었을 때 학습 과정
모델에 데이터(학습데이터) 집어넣어서 추출한 결과를 보는 것

모델을 평가할 때는 모델이 한 번도 학습하지 않은 데이터로 평가해야 한다.
우리가 100장을 준비했다고 치면 일부는 학습 데이터로 쓰고 일부는 검증 데이터로 사용해야 한다.

매번 학습데이터에 대한 Loss와 검증데이터에 대한 Loss를 계산한다.
데이터가 다르기 때문에 좋은 데이터라면 곂칠 수도 있지만 그럴 확률은 낮다.

Train data보다 `validation data의 그래프가 Loss가 높다.`

데이터의 특징을 잘 잡기 시작하는 지점이 있다.
> 어떤 현상이 일어나냐면, train data는 loss가 내려가다가 validation data의 loss는 2차 함수 곡선을 보임(over fitting)
overfitting되기 시작하는 지점에서 멈추면 된다. (가작 epoch이 적으면서 loss가 낮아지는 지점)
애매한게, 어쨌거나 overfitting 되기 시작하는 지점 전에 멈춘다.
underfitting : 검증 데이터만 잘되고 훈련 데이터에 loss가 높다(이런 경우는 거의 없음)
