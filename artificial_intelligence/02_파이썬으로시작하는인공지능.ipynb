## 2.2 두 가지 프로그래밍 환경
- 클라우드 방식
    - 프로그램과 데이터가 서버에 저장되고 관리
        - 서버에 환경이 대부분 갖춰져있음
        - 인터넷만 연결되면 어디서나 개발, 협업 가능
        
    - 구글의 Colab, ~~아마존의 SageMaker~~, 마이크로소프트 Azure
    - 내 프로젝트에 최적인 환경을 갖출 수 없는 한계


- 스탠드얼론 방식
    - 자신에 최적인 환경 구축 가능, 프로그램과 데이터가 자신의 컴퓨터에 저장
    - 소프트웨어를 설치하고 환경을 스스로 구축해야 함.
